# CASE STUDY #05 ANOVA

# XYZ Company has offices in four different zones. 
The company wishes to investigate the following :
● The mean sales generated by each zone.
● Total sales generated by all the zones for each month.
● Check whether all the zones generate the same amount of sales.
Help the company to carry out their study with the help of data provided.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stat
```

```python
data=pd.read_csv('Sales_data_zone_wise.csv')
```

```python
data.head()
```

```python
data.info()
```

# ● The mean sales generated by each zone.

```python
for column in data:
    if(column!="Month"):
        print("Mean sales generated by",column,"is",(data[column].mean()))
```

# ● Total sales generated by all the zones for each month.

```python
data["Total_sales"]=data["Zone - A"]+data["Zone - B"]+data["Zone - C"]+data["Zone - D"]
```

```python
data[['Month','Total_sales']]
```

# ● Check whether all the zones generate the same amount of sales.

H0: Mean sales generated by all the zones are same
H1: Atleast one zone has different mean
alpha = .05
One Way Anova

```python
f_stat, p_value = stat.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
print( 'The f-statistics is : ',f_stat)
print( 'The P value is : ',p_value)
if p_value < .05:
    print("The Null hypothesis is rejected, Which shows the difference in mean sales for atleast one zone.")
else:
    print("The Null Hypothesis is accepted, Which shows that Mean sales generated by all the zones are the same. ")
```
